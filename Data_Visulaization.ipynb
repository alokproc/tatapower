```python
# Import required Library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set a consistent style
sns.set_theme(style="whitegrid")
```

```python
# -----------------------------------------------------------------------------
# 1. Create some synthetic data
# -----------------------------------------------------------------------------
np.random.seed(42)  # For reproducibility

# Example numeric data
x = np.linspace(0, 10, 50)
y = 3 * x + np.random.normal(0, 5, 50)

# Categorical data
categories = np.random.choice(['A', 'B', 'C'], size=50)
values = np.random.randint(1, 20, 50)

# For distribution plots
data_dist = np.random.normal(loc=50, scale=10, size=200)

# For heatmap
matrix_size = 5
data_matrix = np.random.rand(matrix_size, matrix_size)

# Combine into a DataFrame
df = pd.DataFrame({
    'x': x,
    'y': y,
    'category': categories,
    'values': values
})
```

```python
# -----------------------------------------------------------------------------
# 2. Line Chart (Numerical - Numerical, Bivariate)
# -----------------------------------------------------------------------------
plt.figure(figsize=(8,4))
plt.plot(df['x'], df['y'], marker='o')
plt.title('Line Chart')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# -----------------------------------------------------------------------------
# 3. Scatter Chart (Numerical - Numerical, Bivariate)
# -----------------------------------------------------------------------------
plt.figure(figsize=(8,4))
plt.scatter(df['x'], df['y'], c='green')
plt.title('Scatter Chart')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# -----------------------------------------------------------------------------
# 4. Bar Chart (Categorical - Numerical, Bivariate)
# -----------------------------------------------------------------------------
plt.figure(figsize=(8,4))
sns.barplot(data=df, x='category', y='values', estimator=np.mean, ci=None)
plt.title('Bar Chart')
plt.xlabel('Category')
plt.ylabel('Mean Value')
plt.show()
```

```python
# -----------------------------------------------------------------------------
# 5. Pie Chart (Categorical, Univariate)
#    Here, well count how many times each category appears.
# -----------------------------------------------------------------------------
category_counts = df['category'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%')
plt.title('Pie Chart of Category Counts')
plt.show()
```

```python
# -----------------------------------------------------------------------------
# 6. Count Plot (Categorical, Univariate)
# -----------------------------------------------------------------------------
plt.figure(figsize=(8,4))
sns.countplot(data=df, x='category')
plt.title('Count Plot')
plt.show()
```

```python
# -----------------------------------------------------------------------------
# 7. Box Plot (Numerical, Univariate / can be bivariate)
#    Example of box plot by category
# -----------------------------------------------------------------------------
plt.figure(figsize=(8,4))
sns.boxplot(data=df, x='category', y='values')
plt.title('Box Plot')
plt.show()
```

```python
# -----------------------------------------------------------------------------
# 8. Violin Plot (Numerical, Univariate / can be bivariate)
#    Similar idea to box plot but shows distribution shape
# -----------------------------------------------------------------------------
plt.figure(figsize=(8,4))
sns.violinplot(data=df, x='category', y='values')
plt.title('Violin Plot')
plt.show()
```

```python
# -----------------------------------------------------------------------------
# 9. Heatmap (Numerical - Numerical, Bivariate)
#    Using random matrix data
# -----------------------------------------------------------------------------
plt.figure(figsize=(6,5))
sns.heatmap(data_matrix, annot=True, cmap='Blues')
plt.title('Heatmap')
plt.show()
```

```python
# -----------------------------------------------------------------------------
# 10. Histogram (Numerical, Univariate)
# -----------------------------------------------------------------------------
plt.figure(figsize=(8,4))
plt.hist(data_dist, bins=15, color='purple', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# -----------------------------------------------------------------------------
# 11. Distribution Plot (Numerical, Univariate)
# -----------------------------------------------------------------------------
plt.figure(figsize=(8,4))
sns.histplot(data_dist, kde=True, color='orange')
plt.title('Distribution Plot')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```
